# Plot stocks

# 1. Imports

## 1.1 Packages

```python
import os
import pandas as pd

import plotly.graph_objects as go
```

## 1.2 Options

```python
path_data = "../data/01_raw/"
```

## 1.3 Dataset

```python
df_aapl = pd.read_parquet(os.path.join(path_data, "stock_data_aapl.parquet"))
df_msft = pd.read_parquet(os.path.join(path_data, "stock_data_msft.parquet"))
df_amzn = pd.read_parquet(os.path.join(path_data, "stock_data_amzn.parquet"))
df_googl = pd.read_parquet(os.path.join(path_data, "stock_data_googl.parquet"))
df_blk = pd.read_parquet(os.path.join(path_data, "stock_data_blk.parquet"))
```

# 2. Plot stocks

```python
fig = go.Figure(data=[go.Candlestick(
    x=df_aapl.index,
    open=df_aapl['Open']["AAPL"],
    high=df_aapl['High']["AAPL"],
    low=df_aapl['Low']["AAPL"],
    close=df_aapl['Close']["AAPL"]
)])
fig.update_layout(
    title="AAPL Stock Price",
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    xaxis_rangeslider_visible=False
)

fig.show()
```
